```python
plt.scatter(urban_x,urban_y,s=urban_z*smp,facecolors="r",edgecolors ="black",label="Urban")
plt.scatter(suburban_x,suburban_y,s=suburban_z*smp,facecolors="blue",edgecolors ="black",label="Suburban")
plt.scatter(rural_x,rural_y,s=rural_z*smp,facecolors="orange",edgecolors ="black",label="Rural")
plt.xlabel("Ride Count")
plt.ylabel("Average Fare")
plt.title("(*Size of Bubble Depicts Number of Drivers in each City)",fontsize=7)
plt.suptitle("Pyber Ride Sharing Data (2016)", fontsize =15)
plt.legend(loc="best",)
```
